```python
# Let's get the salaries
k = df[df['txn_description'] == 'PAY/SALARY'].sort_values('first_name')
```

```python
# Get distinct salaries for each 
salaries = k.drop_duplicates(subset = ['account'])
salaries.reset_index(drop=True, inplace=True)
salaries['annual_sal'] = salaries.amount * 12
salaries = salaries[['account', 'currency', 'first_name', 'gender', 'age', 'annual_sal']]
```
